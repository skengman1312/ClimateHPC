The geography data come in five resolutions:
* full resolution: Original (full) data resolution.
* high resolution: About 80 % reduction in size and quality.
* intermediate resolution: Another ~80 % reduction.
* low resolution: Another ~80 % reduction.
* crude resolution: Another ~80 % reduction.

### Velocity heat map

```python
# lon_min = f3.variables['lon'][:].min()
# lat_min = f3.variables['lat'][:].min()
# lon_max = f3.variables['lon'][:].max()
# lat_max = f3.variables['lat'][:].max()
m = Basemap(projection='robin',lon_0=0, resolution='c')
# m = Basemap(projection='mill',llcrnrlon=-180,llcrnrlat=-80,urcrnrlon=180,urcrnrlat=90,resolution='i')
```

```python
x,y = m(f3.variables['lon'],f3.variables['lat'])
```

```python
f5.variables['speed'][0]
```

### JANUARY

```python

plt.figure(figsize=(15,15))
plt.tricontourf(x, y, f5.variables['speed'][0],cmap='coolwarm', extend='both')
m.drawcoastlines()
m.drawmapboundary(fill_color='0.9')
m.fillcontinents()
cbar = plt.colorbar(orientation='horizontal', pad=0.03);
cbar.set_label("Speed U")
plt.title('Speed U')
plt.tight_layout()
plt.savefig('velocity_u_january.png', dpi=1000)
```
